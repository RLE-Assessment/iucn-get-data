# EcosystemMap Display Example

This notebook demonstrates how the `EcosystemMap` classes render as HTML tables in Jupyter notebooks.

```python
import ee
from google.auth import default

# Authenticate and initialize Earth Engine
credentials, _ = default(scopes=[
    'https://www.googleapis.com/auth/earthengine',
    'https://www.googleapis.com/auth/cloud-platform'
])
ee.Initialize(credentials=credentials, project='goog-rle-assessments')
```

```python
import pandas as pd
from iucn_get_data import open_ecosystem_map
```

## Load a Raster Ecosystem Asset

When you display an ecosystem map object in a Jupyter cell, it automatically renders as an HTML table showing the asset metadata.

```python
# Define the ecosystem data dictionary for the raster asset
mm_ecosystem_df = pd.DataFrame(
    {
        'get_level3': ['TF1.2', 'T1.1', 'T1.2'],
        'get_level456': ['Myanmar Mangroves', 'Myanmar Lowland Rainforest', 'Myanmar Dry Forest'],
    },
    index=pd.Index([37, 52, 53], name='ecosystem_id')
)

# Load a raster ecosystem asset using asset_id string
raster_ecosystems = open_ecosystem_map(
    'projects/goog-rle-assessments/assets/myanmar/mm_ecosys_v7b',
    ecosystem_band='b1',
    ecosystem_dataframe=mm_ecosystem_df,
)
raster_ecosystems
```

## Load a Vector Ecosystem Asset

Vector assets (FeatureCollections) display feature count and property names.

```python
# Load a vector ecosystem asset using asset_id string with GET Level 3 and Level 4 columns
vector_ecosystems = open_ecosystem_map(
    'projects/goog-rle-assessments/assets/colombia/GETCol',
    get_level3_column='EFG1',
    get_level456_column='Glob_Typol'
)
vector_ecosystems
```

## Create from an ee.FeatureCollection

You can also pass an `ee.FeatureCollection` or `ee.Image` directly to `open_ecosystem_map`.

```python
# Create an ee.FeatureCollection and pass it directly
fc = ee.FeatureCollection('projects/goog-rle-assessments/assets/colombia/GETCol')
ecosystems_from_fc = open_ecosystem_map(
    fc,
    get_level3_column='EFG1',
    get_level456_column='Glob_Typol'
)
ecosystems_from_fc
```

## Access Records as a DataFrame

Use the `dataframe` property to get all records as a pandas DataFrame.

```python
# Get all records as a pandas DataFrame
df = vector_ecosystems.functional_group_dataframe()
df
```
